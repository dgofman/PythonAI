The **softplus** function

```python
import numpy as np
import math

def softplus_np(x): return np.log1p(np.exp(x))
def softplus_math(x): return math.log1p(math.exp(x))
print(softplus_np(2.14))
print(softplus_math(2.14))
print('-------------------------------------')
def softplus_my(x): return log(1 +  math.e ** x)
def log(x):
    n = 104310501
    return n * ((x ** (1 / n)) - 1)
print(softplus_my(2.14))
```

The **sigmoid** function

```python
import numpy as np
import math

def sigmoid_np(x): return 1 / (1 + np.exp(-x))
def sigmoid_math(x): return 1 / (1 + math.exp(-x))
print(sigmoid_np(2.14))
print(sigmoid_math(2.14))
print('-------------------------------------')
def sigmoid_my(x): return 1 / (1 + math.e ** -x)
print(sigmoid_my(2.14))
```

The **ReLU** function

```python
import numpy as np
import math

def relu_np(x): return np.maximum(0, x)
def relu_math(x): return math.max(0, x)
print(relu_np(2.14))
print(relu_np(2.14))
print('-------------------------------------')
def relu_my(x): return x * (x > 0)
print(relu_my(2.14))
```
