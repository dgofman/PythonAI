```python
import numpy as np
import matplotlib.pyplot as plt

N = 100     # number of points
sigma = 20  # standard deviation
y = 2       # y step
```

```python
f = np.array([y * z for z in range(N)])
plt.plot(f)
```

```python
y = f + np.random.normal(0, sigma, N)
x = np.array(range(N))
print("x", x)
plt.scatter(x, y, s=2, c='green')
```

```python
mx = x.sum()/N
print("N:", N, "\nx-sum:", x.sum(), "\nmx:", mx)
```

```python
my = y.sum()/N
print("N:", N, "\ny-sum:", y.sum(), "\nmy:", my)
```

image.png

```python
a2 = np.dot(x.T, x)/N
print("a2:", a2)
```

image.png

```python
a11 = np.dot(x.T, y)/N
print("a11:", a11)
```

image.png

```python
kk = (a11 - mx*my)/(a2 - mx**2)
print("kk:", kk, "\nmx:", mx, "\nmy:", my, "\na2:", a2)
```

image.png

```python
bb = my - kk*mx
print("bb:", bb)
```

image.png

```python
ff = np.array([kk*z+bb for z in range(N)])
plt.plot(ff, c='red')
```

```python
plt.plot(f)
plt.scatter(x, y, s=2, c='green')
plt.plot(ff, c='red')
plt.grid(True)
plt.show()
```
